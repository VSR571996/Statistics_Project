```python

from scipy.stats import ttest_ind, chi2_contingency

# Example: T-test for sales in different regions

a = "sales same in different region"
b = "sales not same in different region"

c =np.array(df[df.Sales == 'yes'].Region)
d =np.array(df[df.Sales == 'no'].Region)
t, p_value = stats.ttest_ind(c,d, axis = 0)
if p_value < 0.05 :
    print(f'{b} as the p_value ({p_value}>0.05')
else :
    print(f'{a} as the p_value ({p_value}<0.05')



# Example: Chi-square test for independence between Ship Mode and Customer Segment



```

```python
o= "Ship mode not independent to Customer Segment"
p= "Ship mode is independent to Customer Segment"

crosstab= pd.crosstab(df['Ship Mode'], df['Customer Segment'])
chi, p_value, dof, expected= stats.chi2_contingency(crosstab)
if p_value<0.5:
    print(f'{o} as the p_value ({p_value.round(3)})<0.5')
else:
    print(f'{p} as the p_value ({p_value.round(3)})>0.5')

```
